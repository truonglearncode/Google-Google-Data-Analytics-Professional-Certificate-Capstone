```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
jan = pd.read_csv('Bike-Share-Data/202401-divvy-tripdata.csv')
feb = pd.read_csv('Bike-Share-Data/202402-divvy-tripdata.csv')
mar = pd.read_csv('Bike-Share-Data/202403-divvy-tripdata.csv')
apr = pd.read_csv('Bike-Share-Data/202404-divvy-tripdata.csv')
may = pd.read_csv('Bike-Share-Data/202405-divvy-tripdata.csv')
jun = pd.read_csv('Bike-Share-Data/202406-divvy-tripdata.csv')
```

```python
jan.head()
```

```python
bike = pd.concat([jan,feb,mar,apr,may,jun])
```

```python
bike.info()
```

```python
bike = bike.drop(['start_station_id','end_station_id','start_lat','start_lng','end_lat','end_lng'], axis = 1)
```

```python
bike.info()
```

```python
bike['started_at'] = pd.to_datetime(bike['started_at'])
bike['ended_at'] = pd.to_datetime(bike['ended_at'])
```

```python
bike.info()
```

```python
bike['hour_of_day'] = bike['started_at'].dt.hour
bike['day_of_week'] = bike['started_at'].dt.day_name()
bike['month_of_year'] = bike['started_at'].dt.month_name()
```

```python
bike.head(5)
```

```python
bike['duration'] = (bike['ended_at'] - bike['started_at']) / pd.Timedelta(minutes = 1)
```

```python
bike['duration'] = bike['duration'].astype('int32')
```

```python
bike.head(5)
```

```python
bike.duplicated(['ride_id']).sum()
```

```python
bike = bike.drop_duplicates(['ride_id'])
```

```python
bike.duplicated(['ride_id']).sum()
```

```python
bike.info()
```

```python
(bike['duration'] <= 0).sum()
```

```python
bike = bike[bike['duration']>0]
```

```python
bike.info()
```

```python
bike.isna().sum()
```

# DEA

### 1. Current differrent of each member? Quantity and Percentage?

```python
bike.head(5)
```

```python
member = bike.groupby('member_casual')[['ride_id']].count()
```

```python
total_rides = member['ride_id'].sum()
```

```python
member['% of member'] = member['ride_id']/ total_rides * 100
```

```python
member
```

```python
plt.figure(figsize = (6,6))
plt.pie(member['ride_id'], labels = member.index, autopct='%.1f%%')
plt.legend(loc = 4)
plt.title('Current different of each member')
plt.show()
```

### 2. Different about using type of bike by each member? Quantity and Percentage?

```python
bike_type = bike.groupby(['member_casual','rideable_type'], as_index = False)[['ride_id']].count()
```

```python
bike_type
```

```python
plt.figure(figsize = (10,5))
sns.barplot(x = 'ride_id', y = 'rideable_type', hue = 'member_casual', data = bike_type)
plt.title('Different about using type of bike by each member')
plt.xlabel('Number of member by each rideable type')
plt.ylabel('Rideable Type')
plt.show()
```

```python
total_count = bike_type.groupby(['rideable_type'])['ride_id'].transform('sum')
```

```python
bike_type['percentage_of_type'] = (bike_type.ride_id/ total_count * 100).round(2)
```

```python
bike_type
```

```python
plt.figure(figsize = (10,5))
sns.barplot(data = bike_type, y = 'rideable_type', x = 'percentage_of_type', hue = 'member_casual')
plt.title('Different about using type of bike by each member')
plt.xlabel('Percentage member by each rideable type')
plt.ylabel('Rideable Type')
plt.show()
```

### 3. How many member of each type using bike service each month? Calculate Percentage for each type of member each month?


```python
member_by_month = bike.groupby(['member_casual','month_of_year'], as_index = False)[['ride_id']].count()
```

```python
total_month_count = member_by_month.groupby(['month_of_year'])['ride_id'].transform('sum')
```

```python
member_by_month['percentage'] = (member_by_month.ride_id/total_month_count *100).round(2)
```

```python
member_by_month
```

```python
plt.figure(figsize = (14,6))
sns.barplot(data = member_by_month, x = 'month_of_year', y = 'percentage', hue = 'member_casual', order = ['January','February','March','April','May','June'])
plt.title('Member of each type using bike service by month')
plt.xlabel('Month of year')
plt.ylabel('Percentage of member using service')
plt.show()
```

### 6. Number of Members Using Bike Service Each Hour of the Day? Percentage of Members Using Bike Service for Each Hour? Overall Percentage for the Entire Day?

```python
bike.head()
```

```python
member_by_hour = bike.groupby(['member_casual','hour_of_day'], as_index = False)[['ride_id']].count()
```

```python
total_hour = member_by_hour.groupby(['hour_of_day'])['ride_id'].transform('sum')
```

```python
member_by_hour['hour_percentage'] = (member_by_hour.ride_id / total_hour *100).round(2)
```

```python
member_by_hour
```

```python
entries_day = member_by_hour.groupby(['member_casual'])['ride_id'].transform('sum')
```

```python
member_by_hour['entries_day_percentage'] = (member_by_hour.ride_id / entries_day *100).round(2)
```

```python
member_by_hour
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize = (16,8))

sns.lineplot(data = member_by_hour, x = 'hour_of_day', y = 'hour_percentage', hue = 'member_casual', ax = ax1)
sns.lineplot(data = member_by_hour, x = 'hour_of_day', y = 'entries_day_percentage', hue ='member_casual', ax = ax2)


ax1.set_title('Percentage of member using bike service by each hour of day')
ax2.set_title('Percentage of member using bike service by entries day')

plt.show()
```
